```python
train["isOpen"] = train.OpenStatus.map({"open": 1, "not a real question": 0, "not constructive": 0, "off topic": 0, "too localized": 0})
```

```python
train["createdAt"] = pd.to_datetime(train.PostCreationDate)
```

```python
train["registeredAt"] = pd.to_datetime(train.OwnerCreationDate)
```

```python
train["ageOfAccount"] = train.createdAt - train.registeredAt
```

```python
train.ageOfAccount.describe()
```

```python
train[train.ageOfAccount < pd.Timedelta(seconds=0)]
```

```python
train["tag_count"] = train[["Tag1", "Tag2", "Tag3", "Tag4", "Tag5"]].isnull().sum(axis=1)
```

```python
train["tag_count"] = (-1*train["tag_count"]) + 5
```

```python
train.columns
```

```python
train["title_length"] = train.Title.apply(len)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9)
```
